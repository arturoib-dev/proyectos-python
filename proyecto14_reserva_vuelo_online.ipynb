# PROYECTO: RESERVA DE VUELOS ONLINE

# Objetivos:
# - Simular la reserva de un vuelo considerando la edad del cliente
# - Determinar el precio base según la clase de vuelo
# - Aplicar descuentos especiales si corresponden
# - Verificar si el saldo disponible es suficiente para completar la reserva
# - Mostrar mensajes claros de éxito o denegación de la reserva

```python
# Proyecto 14: Reserva de vuelo - Método 1

# Datos del cliente
cliente = "Arturo"
edad = 28
clase_vuelo = "SUPERVIP"
saldo_disponible = 200
tiene_descuento = True

# Mostrar información inicial
print("================== RESERVA DE VUELO ===================")
print("Cliente:", cliente)
print("Edad:", edad)
print("Clase de vuelo:", clase_vuelo)
print("Saldo disponible:", round(saldo_disponible, 2), "€")
print("Descuento especial:", "Sí" if tiene_descuento else "No")

# Verificar edad mínima
if edad < 18:
    print("Reserva denegada: Usted es menor de edad")
else:
    # Inicializamos precio_base como None para manejar clases desconocidas
    precio_base = None

    # Asignar precio según clase de vuelo
    if clase_vuelo == "Económica":
        precio_base = 100
    elif clase_vuelo == "Business":
        precio_base = 200
    elif clase_vuelo == "VIP":
        precio_base = 300

# Solo procedemos si precio_base tiene un valor asignado
if precio_base is not None:
    # Aplicar descuento si corresponde
    precio_final = precio_base * 0.8 if tiene_descuento else precio_base

    # Verificar si el cliente tiene saldo suficiente
    if saldo_disponible >= precio_final:
        saldo_restante = saldo_disponible - precio_final
        print("Reserva exitosa con descuento, saldo restante:" if tiene_descuento else
              "Reserva exitosa sin descuento, saldo restante:", round(saldo_restante, 2), "€")
    else:
        dinero_faltante = precio_final - saldo_disponible
        print("Reserva denegada, saldo insuficiente, le falta:", round(dinero_faltante, 2), "€")
else:
    print("Clase de vuelo desconocida")
print("======================================================")
```

```python
# Proyecto 14: Reserva de vuelo - Método 2

# Datos del cliente
cliente = "Arturo"
edad = 28
clase_vuelo = "VIP"
saldo_disponible = 400
tiene_descuento = True

# Mostrar información inicial
print("================== RESERVA DE VUELO ===================")
print("Cliente:", cliente)
print("Edad:", edad)
print("Clase de vuelo:", clase_vuelo)
print("Saldo disponible:", round(saldo_disponible, 2), "€")
print("Descuento especial:", "Sí" if tiene_descuento else "No")

# Verificar edad mínima
if edad < 18:
    print("Reserva denegada: Usted es menor de edad")
else:
    # Verificar si la clase es válida
    if clase_vuelo in ["Económica", "Business", "VIP"]:
        # Asignar precio según clase
        if clase_vuelo == "Económica":
            precio_base = 100
        elif clase_vuelo == "Business":
            precio_base = 200
        else:
            precio_base = 300

        # Aplicar descuento si corresponde
        precio_final = precio_base * 0.8 if tiene_descuento else precio_base

        # Verificar saldo
        if saldo_disponible >= precio_final:
            saldo_restante = saldo_disponible - precio_final
            print("Reserva exitosa con descuento, saldo restante:" if tiene_descuento else
                  "Reserva exitosa sin descuento, saldo restante:", round(saldo_restante, 2), "€")
        else:
            dinero_faltante = precio_final - saldo_disponible
            print("Reserva denegada, saldo insuficiente, le falta:", round(dinero_faltante, 2), "€")
    else:
        print("Clase de vuelo desconocida")
print("======================================================")
```
