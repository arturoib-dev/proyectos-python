# PROYECTO: CONTROL DE ACCESO A UN PARQUE DE DIVERSIONES

# Objetivos:
# - Simular el control de acceso de visitantes a un parque de diversiones
# - Verificar edad y altura mínima para las atracciones
# - Considerar pase VIP para acceso completo
# - Mostrar mensajes personalizados según la situación del visitante

```python
# Proyecto 8: Control de acceso a un parque de diversiones

# ================== DATOS DEL VISITANTE ==================
nombre = "Arturo"
edad = 14
altura = 140  # en cm
tiene_vip = False

# ================== MENSAJES INICIALES ==================
print("=============== CONTROL DE ACCESO ===============")
print("Nombre:", nombre)
print("Edad:", edad)
print("Altura:", altura, "cm")
print("Pase VIP:", "Sí" if tiene_vip else "No")

# ================== LÓGICA DE ACCESO ==================
if edad < 12:
    print("\nNo puedes entrar, edad mínima 12 años")
else:
    if altura >= 130:
        if tiene_vip:
            print("\nAcceso completo a todas las atracciones")
        else:
            print("\nAcceso estándar, algunas atracciones restringidas")
    else:
        print("\nAcceso limitado a atracciones infantiles")

print("=================================================")
```
