# PROYECTO: RESERVA DE HOTEL SEGÚN EDAD Y TIPO DE HABITACIÓN

# Objetivos:
# - Simular la reserva de una habitación de hotel
# - Aplicar descuentos según condiciones
# - Verificar si el cliente tiene suficiente saldo
# - Evaluar restricciones según la edad del cliente
# - Mostrar resultados claros y ordenados

```python
# Proyecto 13: Reserva de hotel según edad y tipo de habitación

# ================== DATOS DEL CLIENTE ==================
cliente = "Arturo"
edad = 28
tipo_habitación = "Individual"
precio_base = 90
tiene_descuento = True
saldo_disponible = 100

# ================== SALIDA INICIAL ==================
print("===================== RESERVA HOTEL =====================")
print("Cliente:", cliente)
print("Edad:", edad)
print("Tipo de habitación:", tipo_habitación)
print("Precio base:", round(precio_base, 2), "€")
print("Descuento aplicado:", "Sí" if tiene_descuento else "No")
print("Saldo disponible", round(saldo_disponible,2), "€")

# ================== LÓGICA DE RESERVA ==================
# Verifica si el cliente es mayor de edad
if edad < 18:
    print("No puede reservar: Usted es menor de edad")
else:
    # Aplica descuento si corresponde
    if tiene_descuento:
        precio_final = precio_base * 0.8  # 20% de descuento
    else:
        precio_final = precio_base

    # Verifica si el saldo disponible alcanza
    if saldo_disponible >= precio_final:
        saldo_restante = saldo_disponible - precio_final
        if tiene_descuento:
            print("Reserva exitosa con descuento, su saldo restante es:", round(saldo_restante, 2), "€")
        else:
            print("Reserva exitosa sin descuento, su saldo restante es:", round(saldo_restante, 2), "€")
    else:
        dinero_faltante = precio_final - saldo_disponible
        print("Reserva denegada, no tiene suficiente saldo, el dinero que le falta es:", round(dinero_faltante, 2), "€")

# ================== SALIDA FINAL ==================
print("=========================================================")
```
