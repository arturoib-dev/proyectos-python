# PROYECTO: VERIFICACIÓN DE INSCRIPCIÓN A UN CURSO ONLINE

# Objetivos:
# - Simular la verificación de inscripción a un curso online
# - Verificar la edad mínima para inscribirse
# - Aplicar descuento si corresponde
# - Mostrar el precio final y recomendar el curso según nivel de conocimientos
# - Proporcionar mensajes claros y organizados al usuario

```python
# Proyecto 9: Verificación de inscripción a un curso online

# ================== DATOS DEL USUARIO ==================
nombre = "Arturo"
edad = 20
nivel_conocimientos = "Intermedio"
precio_curso = 100
tiene_descuento = True

# ================== MENSAJES INICIALES ==================
print("=============== INSCRIPCIÓN CURSO ONLINE ===============")
print("Nombre:", nombre)
print("Edad:", edad)
print("Nivel de conocimientos:", nivel_conocimientos)
print("Precio del curso:", precio_curso)
print("Descuento estudiante:", "Sí" if tiene_descuento else "No")

# ================== LÓGICA DE INSCRIPCIÓN ==================
if edad < 18:
    print("\nNo puede inscribirse, edad mínima 18 años")
else:
    if tiene_descuento:
        precio_final = precio_curso * 0.7  # 30% de descuento
        print("\nPrecio final aplicado con descuento:", round(precio_final, 2))
    else:
        precio_final = precio_curso
        print("\nPrecio final sin descuento:", round(precio_final, 2))

# Recomendación de curso según nivel de conocimientos
print("Curso recomendado:", nivel_conocimientos)

print("========================================================")
```
