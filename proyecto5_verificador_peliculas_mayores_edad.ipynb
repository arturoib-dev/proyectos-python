# PROYECTO: VERIFICADOR  DE PELÍCULAS PARA MAYORES DE EDAD

# Objetivo:
# - Verificar si un usuario puede ver una película según su edad
# - Evaluar si el contenido es apto para menores según el género
# - Aplicar condicionales para permitir o denegar el acceso
# - Mostrar los resultados de forma clara y ordenada

```python
# PROYECTO: VERIFICADOR DE PELÍCULAS PARA MAYORES DE EDAD

# ================== DATOS DEL USUARIO ==================
nombre = "Arturo"
edad = 17
pelicula = "Duro de matar"
genero = "Acción"

# ================== SALIDA INICIAL ==================
print("============= VERIFICANDO ACCESO =============")
print("Nombre:", nombre)
print("Edad:", edad)
print("Película:", pelicula)
print("Género:", genero)

# ================== LÓGICA ==================
# Verifica si el usuario es mayor de edad
if edad >= 18:
    print(nombre, "puede ver cualquier película")

# Si es menor de edad, solo puede ver películas de animación
elif edad < 18 and genero == "Animación":
    print(nombre, "puede ver la película", pelicula)

# En cualquier otro caso, no puede ver la película
else:
    print(nombre, "no puede ver la película", pelicula)

# ================== FINAL ==================
print("==============================================")
```
