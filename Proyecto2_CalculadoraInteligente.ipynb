# PROYECTO: CALCULADORA INTELIGENTE CON OPERACIONES BÁSICAS

# Objetivo:
# - Realizar operaciones matemáticas básicas entre dos números
# - Calcular suma, resta, multiplicación, división y resto
# - Determinar si un número es divisible entre otro
# - Mostrar los resultados de forma ordenada

```python
# Proyecto 2: Calculadora inteligente con operaciones básicas

# ================== DATOS ==================
# Números a operar
num1 = 25
num2 = 4

# ================== CÁLCULOS ==================
# Operaciones matemáticas básicas
suma = num1 + num2
resta = num1 - num2
multiplicacion = num1 * num2
division = num1 / num2
division_entera = num1 // num2
resto = num1 % num2

# Verifica si num1 es divisible entre num2
es_divisible = resto == 0

# ================== SALIDA ==================
print("=========== CALCULADORA INTELIGENTE ===========")
print("Primer número:", num1)
print("Segundo número:", num2)

print("\nResultados:")
print("Suma:", suma)
print("Resta:", resta)
print("Multiplicación:", multiplicacion)
print("División:", round(division, 2))
print("División entera:", division_entera)
print("Resto:", resto)
print("¿Es divisible?", es_divisible)

print("===============================================")
print("Gracias por usar la calculadora")
```
