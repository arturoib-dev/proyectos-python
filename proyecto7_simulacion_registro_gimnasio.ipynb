# PROYECTO: SIMULACIÓN DE REGISTRO EN UN GIMNASIO

# Objetivos:
# - Simular el registro de un usuario en un gimnasio
# - Verificar la edad mínima para registrarse
# - Comprobar si el usuario tiene membresía
# - Evaluar la cantidad de horas semanales y dar recomendaciones
# - Mostrar resultados claros del registro

```python
# Proyecto 7: Simulación de registro en un gimnasio

# ================== DATOS DEL USUARIO ==================
nombre = "Arturo"
edad = 28
tiene_membresía = True
horas_semana = 3

# ================== MENSAJES INICIALES ==================
print("==================== REGISTRO GIMNASIO ====================")
print("Nombre:", nombre)
print("Edad:", edad)
print("Tiene membresía:", "Sí" if tiene_membresía else "No")
print("Horas por semana:", horas_semana)

# ================== LÓGICA DE REGISTRO ==================
if edad < 16:
    print("\nNo puedes registrarte, eres menor de edad")
else:
    if tiene_membresía:
        if horas_semana >= 2:
            print("\nRegistro exitoso, puedes usar todas las instalaciones")
        else:
            print("\nRegistro parcial, se recomienda aumentar actividad")
    else:
        print("\nNecesitas una membresía para registrarte")

print("\nGracias por usar nuestro sistema de registro del gimnasio")
print("===========================================================")
```
