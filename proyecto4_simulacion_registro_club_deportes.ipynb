# PROYECTO: SIMULACIÓN DE REGISTRO EN UN CLUB DE DEPORTES

# Objetivo:
# - Simular el registro de un usuario en un club de deportes
# - Verificar si el usuario tiene la edad mínima requerida
# - Verificar si el usuario tiene membresía
# - Determinar si el acceso al club está permitido o denegado
# - Indicar si el usuario llega tarde al club

```python
# Proyecto 4: Simulación de registro en un club de deportes

# ================== DATOS DEL USUARIO ==================
nombre = "Arturo"
edad = 16
pais = "España"
tiene_membresía = True
hora_llegada = 22  # hora de llegada al club (formato 24h)

# ================== SALIDA ==================
print("========== REGISTRO AL CLUB ==========")
print("Nombre:", nombre)
print("Edad:", edad)
print("País:", pais)
print("Tiene membresía:", tiene_membresía)
print("Hora de llegada:", hora_llegada)

# ================== LÓGICA DE ACCESO ==================
# Verifica si el usuario cumple con la edad mínima y si tiene membresía
if edad >= 18 and tiene_membresía:
    print("\nAcceso permitido al club")
elif edad >= 18 and not tiene_membresía:
    print("Acceso denegado: Necesita membresía")
else:
    print("Acceso denegado: Menor de edad")

# Verifica si el usuario llega tarde
if hora_llegada > 20:
    print("\nLlegaste tarde al club")

print("======================================")
```
