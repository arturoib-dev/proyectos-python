# PROYECTO: RESERVA DE HOTEL CON TEMPORADA, TIPO DE HABITACIÓN Y DESCUENTO

# Objetivos:
# 1. Permitir la reserva de un hotel según la edad del cliente
# 2. Calcular el precio base según el tipo de habitación seleccionada
# 3. Ajustar el precio según la temporada (alta o baja)
# 4. Aplicar descuento especial si corresponde
# 5. Verificar si el cliente tiene saldo suficiente para completar la reserva
# 6. Mostrar mensajes claros de éxito o denegación de la reserva

```python
# Proyecto 15: Reserva de hotel considerando temporada, tipo de habitación y descuento

# ================== DATOS DEL CLIENTE ==================
cliente = "Arturo"
edad = 28
tipo_habitación = "Suite"           # Puede ser Individual, Doble o Suite
temporada = "Alta"                  # Puede ser Alta o Baja
tiene_descuento = True              # True si aplica descuento especial
saldo_disponible = 280              # Saldo disponible del cliente

# ================== INFORMACIÓN INICIAL ==================
print("================== RESERVA DE HOTEL ==================")
print("Cliente:", cliente)
print("Edad:", edad)
print("Tipo de habitación:", tipo_habitación)
print("Descuento especial:", "Sí" if tiene_descuento else "No")
print("Saldo disponible:", saldo_disponible)

# ================== VALIDACIÓN DE EDAD ==================
if edad < 18:
    print("No puede reservar: Usted es menor de edad")
else:
    # ================== VALIDACIÓN TIPO DE HABITACIÓN ==================
    if tipo_habitación in ["Individual", "Doble", "Suite"]:
        # Asignar precio base según el tipo de habitación
        if tipo_habitación == "Individual":
            precio_base = 100
        elif tipo_habitación == "Doble":
            precio_base = 150
        else:  # Suite
            precio_base = 200
        print("Precio base:", round(precio_base, 2), "€")

        # ================== AJUSTE SEGÚN TEMPORADA ==================
        if temporada == "Alta":
            precio_temporada = precio_base + (precio_base * 0.2)
            print("Precio con temporada alta:", round(precio_temporada, 2), "€")
        else:
            precio_temporada = precio_base
            print("Precio con temporada baja:", round(precio_temporada, 2), "€")

        # ================== APLICACIÓN DE DESCUENTO ==================
        if tiene_descuento:
            precio_final = precio_temporada * 0.9  # 10% de descuento
            print("Precio final con descuento:", round(precio_final, 2), "€")
        else:
            precio_final = precio_temporada
            print("Precio final sin descuento:", round(precio_final, 2), "€")

        # ================== VERIFICACIÓN DE SALDO ==================
        if saldo_disponible >= precio_final:
            saldo_restante = saldo_disponible - precio_final
            if tiene_descuento:
                print("Reserva exitosa con descuento, saldo restante:", round(saldo_restante, 2), "€")
            else:
                print("Reserva exitosa sin descuento, saldo restante:", round(saldo_restante, 2), "€")
        else:
            dinero_faltante = precio_final - saldo_disponible
            print("Reserva denegada, saldo insuficiente, le falta:", round(dinero_faltante, 2), "€")
    else:
        print("Tipo de habitación no válida")

print("=====================================================")
```
