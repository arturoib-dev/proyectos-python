# PROYECTO: COMPRA ONLINE DE ENTRADAS PARA UN CONCIERTO

# Objetivos:
# - Simular la compra online de entradas para un concierto presencial
# - Verificar la edad mínima del cliente
# - Aplicar descuentos si corresponden
# - Comprobar si el saldo disponible es suficiente
# - Mostrar el saldo restante o la cantidad faltante
# - Proporcionar mensajes claros y organizados al usuario

```python
# Proyecto 10: Simulación de compra online de entradas para un concierto

# ================== DATOS DEL CLIENTE ==================
cliente = "Arturo"
edad = 28
tipo_entrada = "VIP"
precio_base = 105.00
tiene_descuento = True
saldo = 120

# ================== MENSAJES INICIALES ==================
print("==================== CONCIERTO ====================")
print("Cliente:", cliente)
print("Edad:", edad)
print("Tipo de entrada:", tipo_entrada)
print("Precio base:", round(precio_base, 2), "€")
print("Descuento especial:", "Sí" if tiene_descuento else "No")
print("Saldo disponible:", round(saldo, 2), "€")

# ================== LÓGICA DE COMPRA ==================
if edad < 18:
    print("\nNo puedes comprar entradas, debes ser mayor de edad")
else:
    # Aplicar descuento si corresponde
    if tiene_descuento:
        precio_final = precio_base * 0.85  # 15% de descuento
    else:
        precio_final = precio_base

    # Verificar saldo disponible
    if saldo >= precio_final:
        saldo_restante = saldo - precio_final
        if tiene_descuento:
            print("\nCompra exitosa online, con descuento. Saldo restante:", round(saldo_restante, 2), "€")
        else:
            print("\nCompra exitosa online. Saldo restante:", round(saldo_restante, 2), "€")
    else:
        dinero_faltante = precio_final - saldo
        print("\nSaldo insuficiente, te falta:", round(dinero_faltante, 2), "€")

print("===================================================")
```
