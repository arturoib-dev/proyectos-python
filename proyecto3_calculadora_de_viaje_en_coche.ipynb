#PROYECTO: CALCULADORA DE GASTOS DE VIAJE CON CONTROL DE PRESUPUESTO

# Objetivo:
# - Calcular la duración estimada de un viaje
# - Calcular el combustible necesario para el viaje
# - Calcular el costo total del combustible
# - Evaluar si el presupuesto disponible es suficiente
# - Determinar si el viaje es largo o costoso

```python
# Proyecto 3: Calculadora de gastos de viaje con control de presupuesto

# ================== DATOS DEL VIAJE ==================
distancia = 450             # km
velocidad_promedio = 90     # km/h
consumo_cada_100_km = 6.5  # litros cada 100 km
precio_del_litro = 1.8      # € por litro
presupuesto_disponible = 120  # €

# ================== CÁLCULOS ==================
# Duración estimada del viaje en horas
duracion_estimada = distancia / velocidad_promedio

# Cantidad de combustible necesario en litros
combustible_necesario = (distancia / 100) * consumo_cada_100_km

# Costo total del combustible
costo_total = combustible_necesario * precio_del_litro

# ================== BOOLEANOS ==================
# Verifica si el presupuesto es suficiente
presupuesto_suficiente = presupuesto_disponible >= costo_total

# Determina si el viaje es largo (>300 km)
viaje_largo = distancia > 300

# Determina si el viaje es costoso (>100 €)
viaje_costoso = costo_total > 100

# ================== SALIDA ==================
print("=========== CALCULADORA DE VIAJE ===========")
print("Distancia:", distancia, "km")
print("Velocidad Promedio:", velocidad_promedio, "km/h")
print("Consumo cada 100 km:", consumo_cada_100_km, "litros")
print("Precio del litro:", precio_del_litro, "€")
print("Presupuesto disponible:", presupuesto_disponible, "€\n")

print("Duración estimada:", round(duracion_estimada, 2), "horas")
print("Combustible necesario:", round(combustible_necesario, 2), "litros")
print("Costo total:", round(costo_total, 2), "€\n")

print("¿Presupuesto suficiente?", presupuesto_suficiente)
print("¿Viaje largo?", viaje_largo)
print("¿Viaje costoso?", viaje_costoso)
print("============================================")
print("           ¡Buen viaje, conductor!")
```
