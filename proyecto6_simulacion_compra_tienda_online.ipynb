# PRYECTO: SIMULACIÓN DE COMPRA EN UNA TIENDA ONLINE

# Objetivos:
# - Simular el proceso de compra de un cliente en una tienda online
# - Aplicar descuentos mediante cupones
# - Verificar si el cliente tiene saldo suficiente
# - Mostrar resultados claros de la transacción

```python
# Proyecto 6: Simulación de compra en una tienda online

# ================== DATOS DEL CLIENTE ==================
nombre_cliente = "Arturo"
edad = 28
producto = "Auriculares"
precio = 50.0
cupon_descuento = True
saldo = 60.0

# ================== MENSAJES INICIALES ==================
print("========== TIENDA ONLINE ==========")
print("Cliente:", nombre_cliente)
print("Edad:", edad)
print("Producto:", producto)
print("Precio:", precio, "€")
print("Cupón de descuento:", "Sí" if cupon_descuento else "No")
print("Saldo disponible:", saldo, "€")

# ================== LÓGICA DE COMPRA ==================
if edad < 18:
    print("\nNo puedes comprar este producto")
else:
    # Aplicar descuento si hay cupón
    if cupon_descuento:
        precio_final = precio * 0.8  # 20% de descuento
    else:
        precio_final = precio

    # Verificar si el saldo alcanza para la compra
    if saldo >= precio_final:
        saldo_restante = saldo - precio_final
        if cupon_descuento:
            print("\nCompra exitosa con descuento, tu saldo restante es:", round(saldo_restante, 2), "€")
        else:
            print("\nCompra exitosa, tu saldo restante es:", round(saldo_restante, 2), "€")
    else:
        dinero_faltante = precio_final - saldo
        print("\nSaldo insuficiente, necesitas", round(dinero_faltante, 2), "€ más")

print("===================================")
```
