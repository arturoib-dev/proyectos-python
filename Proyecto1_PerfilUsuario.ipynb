# PROYECTO: PERFIL DE USUARIO CON CÁLCULO DE EDAD E IMC

# Objetivo:
# - Crear un perfil de usuario
# - Calcular año de nacimiento
# - Calcular IMC
# - Evaluar condiciones básicas

```python
# Ejercicio: Perfil de usuario con cálculo de edad e IMC

# ================== DATOS DEL USUARIO ==================
# Información básica del usuario
nombre = "Arturo"
apellido = "Iparraguirre"
edad = 28
altura = 1.75  # en metros
peso = 80.5    # en kg
pais = "España"
ciudad = "Madrid"

# ================== CÁLCULOS ==================
# Cálculo del año de nacimiento (aproximado)
año_actual = 2026
año_de_nacimiento = año_actual - edad - 1

# Cálculo del Índice de Masa Corporal (IMC)
imc = peso / (altura ** 2)

# ================== BOOLEANOS ==================
# Verifica si el usuario es mayor de edad
es_mayor_de_edad = edad >= 18

# Verifica si el IMC está en rango saludable
imc_saludable = imc >= 18.5 and imc <= 24.9

# ================== SALIDA ==================
# Muestra la información del usuario de forma ordenada

print("========== PERFIL DEL USUARIO ==========")
print("Nombre completo:", nombre, apellido)
print("Ubicación:", ciudad, ",", pais)
print("Edad:", edad, "años")
print("Año de nacimiento:", año_de_nacimiento)

print("\n============ DATOS FÍSICOS ============")
print("Altura:", altura, "m", "\tPeso:", peso, "kg")
print("IMC:", round(imc, 2))

print("\n¿Es mayor de edad?", es_mayor_de_edad)
print("¿Tiene IMC saludable?", imc_saludable)

print("\n=======================================")
print("Gracias por usar el programa,", nombre, "!")
```
