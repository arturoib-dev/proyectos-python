# PROYECTO: RECOMENDACIONES PARA SALIR SEGÚN CLIMA Y ACTIVIDAD

# Objetivos:
# - Recibir información del usuario: nombre, clima, temperatura y tipo de actividad
# - Dar recomendaciones según la temperatura
# - Dar recomendaciones según el clima
# - Dar recomendaciones según la actividad
# - Mostrar toda la información de manera organizada y clara

```python
# Proyecto 11: Recomendaciones según clima y actividad

# ================== DATOS DEL USUARIO ==================
nombre = "Arturo"
clima = "Lluvioso"
temperatura = 22  # en °C
actividad = "Deporte"

# ================== SALIDA ==================
print("===== RECOMENDACIÓN PARA EL USUARIO =====")
print("Nombre:", nombre)
print("Clima:", clima)
print("Temperatura:", temperatura, "°C")
print("Actividad:", actividad)
print("\nRecomendación:")

# Recomendaciones según la temperatura
if temperatura >= 30:
    print("Hace mucho calor, hidrátate")
elif temperatura < 15:
    print("Hace mucho frío, abrígate")
else:
    print("Temperatura agradable")

# Recomendaciones según el clima
if clima == "Soleado":
    print("Lleva gafas de sol")
elif clima == "Nublado":
    print("Lleva ropa cómoda")
else:
    print("Lleva impermeable y paraguas")

# Recomendaciones según la actividad
if actividad == "Deporte":
    print("Usa ropa cómoda")
elif actividad == "Paseo":
    print("Usa ropa casual")
else:
    print("Usa ropa cómoda para casa")

print("=========================================")
```
