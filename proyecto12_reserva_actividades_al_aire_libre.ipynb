# PROYECTO: RESERVA DE ACTIVIDADES AL AIRE LIBRE

# Objetivos:
# - Simular la reserva de actividades al aire libre
# - Evaluar condiciones climáticas y temperatura para decidir si se puede realizar la actividad
# - Verificar si el usuario cuenta con el equipo adecuado
# - Mostrar recomendaciones basadas en la combinación de clima, temperatura y equipo

```python
# Proyecto 12: Reserva de activiades al aire libre - Método 1

# Datos del usuario
nombre = "Arturo"
temperatura = 22
clima = "Frío"
actividad = "Deporte"
equipo_adecuado = True

# Mostrar información del usuario y actividad
print("================ RESERVA DE ACTIVIDAD ================")
print("Nombre:", nombre)
print("Temperatura:", temperatura)
print("Clima:", clima)
print("Actividad:", actividad)
print("Equipo adecuado:", "Sí" if equipo_adecuado else "No")

# Lógica de recomendación según temperatura y clima
if temperatura < 15:
    print("Recomendación: Demasiado frío, no se puede hacer la actividad")
else:
    if clima == "Soleado":
        if equipo_adecuado:
            print("Recomendación: Puedes realizar la actividad con éxito")
        else:
            print("Recomendación: No puedes realizar la actividad con éxito")
    elif clima == "Nublado":
        if equipo_adecuado:
            print("Recomendación: Puedes realizar la actividad con éxito")
        else:
            print("Recomendación: No puedes realizar la actividad con éxito")
    elif clima == "Lluvioso":
        if equipo_adecuado:
            print("Recomendación: Puedes realizar la actividad con éxito")
        else:
            print("Recomendación: No puedes realizar la actividad con éxito")
    else:
        print("Clima desconocido, verifica los datos")
print("======================================================")
```
